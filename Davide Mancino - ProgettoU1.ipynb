```python
import statsmodels.api as sm
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.tsa.stattools import adfuller, kpss
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.deterministic import DeterministicProcess

import warnings
```

```python
def adf_test(timeseries):
    print ('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print (dfoutput)
```

```python
def kpss_test(timeseries):
    print ('Results of KPSS Test:')
    kpsstest = kpss(timeseries, regression='c')
    kpss_output = pd.Series(kpsstest[0:3], index=['Test Statistic','p-value','Lags Used'])
    for key,value in kpsstest[3].items():
        kpss_output['Critical Value (%s)'%key] = value
    print (kpss_output)
```

# UCM

```python
def UCM_model(lev_tre, train, validation, covid):
    if covid==0:
        model = sm.tsa.UnobservedComponents(train['CONSUMO_ATTIVA_PRELEVATA'],
                                            lev_tre,
                                            seasonal=24, freq_seasonal = [{'period':24*7, 'harmonics':15},{'period':24*365, 'harmonics':5}])
        model = model.fit(low_memory=True)
        predicted_val = model.get_prediction(start=validation.index[0], end=validation.index[-1])
        predicted_val_mean = predicted_val.predicted_mean
        residui_val=validation['CONSUMO_ATTIVA_PRELEVATA']-predicted_val_mean   
  
  
    elif covid==1:
        model = sm.tsa.UnobservedComponents(train['CONSUMO_ATTIVA_PRELEVATA'],
                                            lev_tre,
                                            exog=np.array(df_u1_info['covid'])[:len(train)],
                                            seasonal=24, freq_seasonal = [{'period':24*7, 'harmonics':15},{'period':24*365, 'harmonics':5}])  
        model = model.fit()
        predicted_val = model.get_prediction(start=validation.index[0], end=validation.index[-1], 
                                            exog=np.array(df_u1_info['covid'])[len(train):,].reshape(len(validation), 1))
        predicted_val_mean = predicted_val.predicted_mean
        residui_val=validation['CONSUMO_ATTIVA_PRELEVATA']-predicted_val_mean
  
    min=model.loglikelihood_burn
    predicted_train = model.get_prediction(start=train.index[min], end=train.index[-1])
    predicted_train_mean = predicted_train.predicted_mean
    residui_train=train['CONSUMO_ATTIVA_PRELEVATA'][min:]-predicted_train_mean

    print("level-trend: {}" .format(lev_tre))
    print("MAE train:", mean_absolute_error(train['CONSUMO_ATTIVA_PRELEVATA'][min:], predicted_train_mean))
    print("MAE validation:", mean_absolute_error(validation['CONSUMO_ATTIVA_PRELEVATA'], predicted_val_mean), '\n')
```
